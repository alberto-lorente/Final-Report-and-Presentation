```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import json
import os
```

#Loading the results dictionaries and merging them into a single dataframe

```python


train = r"Models Performance\Training Celebrity dataset"
test = r"Models Performance\Testing Fake News dataset"

list_test = os.listdir(test)
list_train = os.listdir(train)

```

```python
list_dicts_train = []

for file in list_train:
    
    path = str(train)+ "\{}".format(file)
    
    # print(path)
    
    with open(path) as f:
        data = json.load(f)
        list_dicts_train.append(data)
        
list_dicts_test = []

for file in list_test:
    
    path = str(test)+ "\{}".format(file)
    
    # print(path)
    
    with open(path) as f:
        data = json.load(f)
        list_dicts_test.append(data)
        
```

```python
results_list = list_dicts_test + list_dicts_train 
```

```python
print(len(results_list))
```

```python
results_df = pd.DataFrame(results_list)
# results_df.to_csv("results.csv")
```
