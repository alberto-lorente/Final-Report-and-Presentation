```python
list_dicts_model_yes = []

for file in list_model_yes:
    
    path = str(model_yes)+ "\{}".format(file)
    
    # print(path)
    
    with open(path) as f:
        data = json.load(f)
        list_dicts_model_yes.append(data)
```
